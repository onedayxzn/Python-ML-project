```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Pokemon.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df[df['Name']=='Jolteon']
```

```python
effective_list = ['Grass', 'Electric', 'Dragon', 'Ground']

df_type = df[df['Type 1'].isin(effective_list)]

df_type.head()
```

```python
df_type_gen1 = df_type[df_type['Generation']==1]
df_type_gen1
```

```python
df_type_gen1_sorted = df_type_gen1.sort_values(by=['Sp. Def'], ascending=False)
df_type_gen1_sorted
```

```python
sns.boxplot(x='Generation', y="HP", data=df)
```

```python
quantile_97 = np.quantile(df["HP"], 0.97)
quantile_97
```

```python
hp_outliers_df = df[df['HP'] > quantile_97]
hp_outliers_df
```

```python
hp_outliers_id = hp_outliers_df.index.tolist()
hp_outliers_id
```

```python
df_hp_no_outliers = df.drop(labels=hp_outliers_id)
df_hp_no_outliers
```

```python
print(df.shape)
print(df_hp_no_outliers.shape)
```

```python
df_type_gen1_sorted = df_type_gen1.sort_values(by = 'Sp. Def', axis=0, ascending=False)
df_type_gen1_sorted.head()
```
